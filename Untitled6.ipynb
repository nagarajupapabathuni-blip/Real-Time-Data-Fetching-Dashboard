```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import time

# ----------------------------------------
# REAL TIME WEATHER DATA FETCHING PROJECT
# ----------------------------------------

# Empty list to store data
weather_records = []

# API URL (Vijayawada Coordinates)
API_URL = "https://api.open-meteo.com/v1/forecast?latitude=16.52&longitude=80.62&current_weather=true"

print("Fetching Real Time Weather Data...\n")

# Fetch data multiple times
for i in range(5):

    print(f"Fetching Record {i+1} ...")

    # API Request
    response = requests.get(API_URL)

    # Convert response into JSON
    data = response.json()

    # Extract current weather information
    current_weather = data['current_weather']

    temperature = current_weather['temperature']
    windspeed = current_weather['windspeed']
    winddirection = current_weather['winddirection']
    weathercode = current_weather['weathercode']
    fetch_time = current_weather['time']

    # Current system time
    system_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    # Store data in dictionary
    weather_info = {
        "System_Time": system_time,
        "API_Time": fetch_time,
        "Temperature": temperature,
        "Wind_Speed": windspeed,
        "Wind_Direction": winddirection,
        "Weather_Code": weathercode
    }

    # Append data
    weather_records.append(weather_info)

    print("Temperature:", temperature)
    print("Wind Speed:", windspeed)
    print("Wind Direction:", winddirection)
    print("Weather Code:", weathercode)
    print("-" * 50)

    # Wait for 2 seconds before next fetch
    time.sleep(2)

# ----------------------------------------
# CREATE DATAFRAME
# ----------------------------------------

weather_df = pd.DataFrame(weather_records)

print("\nFinal Weather Data:\n")
print(weather_df)

# ----------------------------------------
# SAVE CSV FILE
# ----------------------------------------

weather_df.to_csv("real_time_weather_data.csv", index=False)

print("\nCSV File Saved Successfully")

# ----------------------------------------
# DATA ANALYSIS
# ----------------------------------------

average_temp = weather_df['Temperature'].mean()
max_temp = weather_df['Temperature'].max()
min_temp = weather_df['Temperature'].min()

average_wind = weather_df['Wind_Speed'].mean()

print("\nWeather Analysis")
print("-" * 30)

print("Average Temperature:", average_temp)
print("Maximum Temperature:", max_temp)
print("Minimum Temperature:", min_temp)

print("Average Wind Speed:", average_wind)

# ----------------------------------------
# VISUALIZATION 1
# TEMPERATURE GRAPH
# ----------------------------------------

plt.figure(figsize=(10,5))

plt.plot(
    weather_df['System_Time'],
    weather_df['Temperature'],
    marker='o'
)

plt.title("Real Time Temperature Monitoring")
plt.xlabel("Time")
plt.ylabel("Temperature")

plt.xticks(rotation=45)

plt.tight_layout()

plt.savefig("temperature_graph.png")

plt.show()

# ----------------------------------------
# VISUALIZATION 2
# WIND SPEED GRAPH
# ----------------------------------------

plt.figure(figsize=(10,5))

plt.bar(
    weather_df['System_Time'],
    weather_df['Wind_Speed']
)

plt.title("Wind Speed Analysis")
plt.xlabel("Time")
plt.ylabel("Wind Speed")

plt.xticks(rotation=45)

plt.tight_layout()

plt.savefig("windspeed_graph.png")

plt.show()

# ----------------------------------------
# VISUALIZATION 3
# PIE CHART
# ----------------------------------------

labels = ['Average Temp', 'Average Wind']

values = [average_temp, average_wind]

plt.figure(figsize=(7,7))

plt.pie(
    values,
    labels=labels,
    autopct='%1.1f%%'
)

plt.title("Weather Summary Distribution")

plt.savefig("weather_summary_piechart.png")

plt.show()

# ----------------------------------------
# FINAL REPORT GENERATION
# ----------------------------------------

report = f"""
======================================
REAL TIME WEATHER DATA FETCHING REPORT
======================================

Total Records Collected : {len(weather_df)}

Average Temperature : {average_temp}
Maximum Temperature : {max_temp}
Minimum Temperature : {min_temp}

Average Wind Speed : {average_wind}

Files Generated:
1. real_time_weather_data.csv
2. temperature_graph.png
3. windspeed_graph.png
4. weather_summary_piechart.png

Project Status : SUCCESS
"""

# Save report
with open("project_report.txt", "w") as file:
    file.write(report)

print(report)

print("Project Completed Successfully")
```
